```python
data.drop(['Car_Name'],axis=1,inplace=True)
data
```

```python
data["Cur_Year"]=2021
data
```

```python
data['No_of_Years']=data['Cur_Year']-data['Year']
data
```

```python
data.drop(['Year'],axis=1,inplace=True)
data.drop(['Cur_Year'],axis=1,inplace=True)
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
#independent and dependend features
x=data.drop(['Selling_Price'],axis=1)
y=data['Selling_Price']
print(x.head())
```

```python
y.head()
```

```python
#feature importance
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import math
list=[100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500]
mini=100000000000
n=-1
for i in list:
    rf=RandomForestRegressor(n_estimators=i,criterion='mse')
    rf.fit(xtrain,ytrain)
    kf=rf.predict(xtest)
    if mini>math.sqrt(mean_squared_error(ytest,kf)):
        mini=math.sqrt(mean_squared_error(ytest,kf))
        n=i;
print("min is",n,mini)  
```

```python
rf=RandomForestRegressor(n_estimators=n,criterion='mse')
rf.fit(xtrain,ytrain)
pred=rf.predict(xtest)
```

```python
sns.distplot(ytest-pred)
```

```python
sns.scatterplot(ytest,pred)
```

```python
import pickle
#open a file where u want to store the data
file=open("car_price_predictor.pkl",'wb')
pickle.dump(rf,file)
```
